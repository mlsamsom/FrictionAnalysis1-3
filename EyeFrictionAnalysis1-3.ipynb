# Data IO and parsing
The file name is read in using a gui utility found in the TKinter library. Raw data is parsed based on the project file: filename.prj. Please ensure that this is the correct usage. 

```python
# GUI to choose file
Tk().withdraw() # we don't want a full GUI, so keep the root window from appearing
fname = askopenfilename() # show an "Open" dialog box and return the path to the selected file
print 'file chosen: ', fname

# get length of csv file (can't do this in the other block for some reason)
f = open(fname, 'r')
sampler = csv.reader(f)
row_count = sum(1 for row in sampler)
f.close()

f = open(fname, 'r')
reader = csv.reader(f)
rawdata = np.zeros((8,row_count))
for row in reader:
    rownum = int(reader.line_num)-1
    try:
        indata = np.asarray(row)
        indata[indata==''] = 0.
        rawdata[:,rownum] = indata.astype(np.float)
    except:
        headers = row
f.close()

#ID data blocks
tare =   (0, 5, 10)
ramp =   (1, 6, 11)
dwell =  (2, 7, 12)
run_pos = (3, 8, 13)
run_neg = (4, 9, 14)

block = 0
scount = 0
ecount = 0 
bstart = np.zeros(15)
bend = np.zeros(15)
for i in xrange(row_count-2):
    if rawdata[0,i] == 0 and rawdata[0,i+1] != 0:
        bstart[scount] = i+1
        scount+=1
    elif rawdata[0,i] != 0 and rawdata[0,i+1] == 0:
        bend[ecount] = i+1
        ecount+=1
bend[14] = row_count # fill in last index
```

# Data analysis
Data analysis is below. This data analysis does not look at the available ramp compression of the samples, this could be added.<br><br> This data analysis calculates pressure as:
$$ P = N_{tavg}/A $$
where, 
$$ P = pressure $$   
$$ N_{tavg} = average\, load\, of\, the\, final\, two\, rotations\, of\, both\, positive\, and\, negative\, directions $$ 
$$ A = area\, of\, annulus $$
Static friction coefficient:
$$ \mu_{static} = \frac{\tau^{+}_{max} - \tau^{-}_{max}} {N_{dwell} \cdot R_{eff}}$$ 
Where 
$$ \tau_{max} = max\, torque\, in\, a\, given\, direction\, measured\, during\, the\, first\, 20^{\circ}\, rotation $$ 
$$ N_{dwell} = average\, dwell\, $$ 
$$ R_{eff} = 2.44mm $$
Kinetic friction coefficient is given by:
$$ \mu_{kin} = \frac{ \tau^{+}_{avg} }{ N^{+}_{avg} \cdot R_{eff} } - \frac{ \tau^{-}_{avg} }{ N^{-}_{avg} \cdot R_{eff} } $$
where, 
$$ \tau_{avg} = average\, torque\, over\, the\, final\, 2\, rotations\, of\, a\, given\, direction $$
$$ N_{avg} = average\, load\, over\, the\, final\, 2\, rotations\, of\, a\, given\, direction $$

```python
#ANALYSIS IS DONE HERE#################################################
#######################################################################
#Processing data
#  Channels: 
#   Data points: 0
#   Elapsed time: 1
#   Scan time: 2
#   Disp: 3
#   Load: 4
#   Torque: 5
#   Rotation: 6

# CONSTANTS
Reff = 0.00244 # (m)
AREA = 24.6 #(mm**2)
samples = 2048 # number of points samples during rotation

#  allocating and analyze data
tarevals = np.zeros((2,3)) # load and torque, run 1, 2, 3
init_load = np.zeros(3)    # mean initial load
runs = np.zeros((2,3,int(samples*2)))

mu_stat = np.zeros(3)
mu_kin = np.zeros(3)
P = np.zeros(3)

for i in xrange(3):
    tarevals[0,i] = np.mean(rawdata[4, bstart[tare[i]]:bend[tare[i]] ]) #tare load
    tarevals[1,i] = np.mean(rawdata[5, bstart[tare[i]]:bend[tare[i]] ]) #tare torque
    init_load[i] = np.mean(rawdata[ 4, bstart[dwell[i]]:bend[dwell[i]] ]) #dwell load
    
    runs[0,i,:samples] = rawdata[4, bstart[run_pos[i]]:bend[run_pos[i]] ] #load
    runs[0,i,samples:] = rawdata[4, bstart[run_neg[i]]:bend[run_neg[i]] ] #load
    runs[1,i,:samples] = rawdata[5, bstart[run_pos[i]]:bend[run_pos[i]] ] #torque
    runs[1,i,samples:] = rawdata[5, bstart[run_neg[i]]:bend[run_neg[i]] ] #torque
    
    #tare
    runs[0,i,:] = runs[0,i,:] - tarevals[0,i]    
    runs[1,i,:] = runs[1,i,:] - tarevals[1,i]
    init_load[i] = init_load[i] - tarevals[0,i]
    
    # static friction coefficient (max first 20 deg rotation)
    tm_pos = np.max(runs[1,i,:30])
    tm_neg = np.max([runs[1,i,samples:int(samples+30)]])
    N_init = np.absolute(init_load[i])
    mu_stat[i] = (tm_pos - tm_neg) / (N_init * Reff * 2)
    mu_stat[i] = np.absolute(mu_stat[i]) # make sure it's positive
    
    # kinetic friction coefficient (mean of final 2 rotations)
    N_pos = np.mean(runs[0,i,int(samples/2):samples]) 
    N_pos = np.absolute(N_pos)
    t_pos = np.mean(runs[1,i,int(samples/2):samples])
    N_neg = np.mean(runs[0,i,int(samples + samples/2):])
    N_neg = np.absolute(N_neg)
    t_neg = np.mean(runs[1,i,int(samples + samples/2):])
    
    mu_kin[i] = 2 * ( ( t_pos/(N_pos*Reff) ) - ( t_neg/(N_neg*Reff) ) )
    mu_kin[i] = np.absolute(mu_kin[i]) # make sure it's positive
    
    P[i] = (N_pos + N_neg) / 2
    P[i] = (P[i] / AREA)*1000
    
print "Pressure (kPa):", P   
print "u static:", mu_stat
print "u kinetic:", mu_kin
```

# Output
Outputs pressure, static and kinetic friction values into a .csv file titled: FrictionAnalysis_output.csv

```python
# Build output file
with open('FrictionAnalysis_output.csv', 'wb') as csvfile:
    print 'building .csv output file \FrictionAnalysis_output.csv'
    writer = csv.writer(csvfile)
    writer.writerow(['','Load 1','Load 2','Load 3'])
    writer.writerow(['Pressure', P[0], P[1], P[2]])
    writer.writerow(['stiction', mu_stat[0], mu_stat[1], mu_stat[2]])
    writer.writerow(['friction', mu_kin[0], mu_kin[1], mu_kin[2]])
csvfile.close()
```
